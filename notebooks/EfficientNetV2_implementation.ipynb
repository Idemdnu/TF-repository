```python
def load_and_resize_images(folder_path, target_size=(224, 224), show_images=False):
    """
    """
    images = []
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            with Image.open(file_path) as img:
                img = img.resize(target_size, Image.ANTIALIAS)
                img = img.convert('RGB')
                image_array = np.array(img)

            #Prepocessing using tensorflow

                image_array = preprocess_input(image_array)
                images.append(image_array)
                if show_images:
                    plt.imshow(image_array)
                    plt.title(f"Resized: {filename}")
                    plt.axis('off')  # Turn off axis numbers and ticks
                    plt.show()
    return np.array(images)
```

```python
#https://chatgpt.com/c/e84d39d1-96c6-49ec-bdcd-a998d5948d33

# Load your pre-trained TensorFlow model (e.g., EfficientNet)
model = tf.keras.models.load_model('C:\\Users\\Patrik Korsch\\Desktop\\VSE\DAB\\Python\\tensorflow_computer_vision_project\\EffNetV2B0.keras')

def load_and_resize_image(uploaded_file, target_size=(224, 224), show_image=False):
    """
    Load an uploaded image file, resize it, and preprocess it for model prediction.
    
    Parameters:
        uploaded_file: The image file uploaded by the user.
        target_size (tuple): The target size to which the image is resized.
        show_image (bool): If True, display the image using matplotlib.
    
    Returns:
        np.array: A numpy array of the preprocessed image.
    """
    # Open the image file and resize it
    img = Image.open(uploaded_file)
    img = img.resize(target_size, Image.ANTIALIAS)
    img = img.convert('RGB')  # Convert the image to RGB format
    image_array = np.array(img)

    # Preprocess the image using the specific model's preprocessing function
    image_array = preprocess_input(image_array)

    # Optionally show the image
    if show_image:
        plt.imshow(image_array)
        plt.title("Processed Image")
        plt.axis('off')
        plt.show()

    # Since the model expects a batch dimension, expand dimensions to simulate a batch
    return np.expand_dims(image_array, axis=0)

def predict(image):
    # Predict the class of the image
    prediction = model.predict(image)
    return prediction

st.title('Image Classification with EfficientNet')

uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])
if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption='Uploaded Image', use_column_width=True)
    st.write("")
    st.write("Classifying...")
    # Preprocess the image and predict
    processed_image = load_and_resize_image(uploaded_file, show_image=False)
    prediction = predict(processed_image)
    st.write(f'Prediction: {prediction}')
```
